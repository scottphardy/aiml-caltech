```python
# Import required libraries

import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
df = pd.read_csv('NSMES1988.csv')
```

```python
# Convert numerical columns to int16
df_int_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[df_int_cols] = df[df_int_cols].astype('int16')
```

```python
df.info(memory_usage='deep')
```

```python
df.isna().sum()
```

```python
df['age'] = df['age'] * 10
```

```python
df['income'] = df['income'] * 10000
```

```python
df['age'].value_counts(bins=10)
```

```python
df.to_csv('income.csv', sep='|', index=False)
```

```python
df.head()
```

```python
total_visits = df[['visits', 'nvisits', 'ovisits', 'novisits', 'emergency', 'hospital']].sum(axis=1)
```
